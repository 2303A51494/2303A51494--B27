```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("GlobalCO2Emissions.csv")
print(df.head())
plt.figure(figsize=(10,6))
plt.plot(df["Year"], df["Emissions"], marker='o', linestyle='-', color='b')
plt.xlabel("Year")
plt.ylabel("CO2 Emissions (in million metric tons)")
plt.title("Global CO2 Emissions Over Time")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_scores.csv")

plt.figure(figsize=(8,5))
plt.plot(df["Exam"], df["Scores"], marker='o', linestyle='-', color='g')
plt.title("Student Performance Across Exams")
plt.xlabel("Exam Number")
plt.ylabel("Score (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
