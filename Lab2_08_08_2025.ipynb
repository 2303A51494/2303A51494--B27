```python
import numpy as np
import pandas as pd
arr=np.array([10,20,30,40,50])
print('Original Array:' ,arr)
print('Add 5 to each element' ,arr+5)
print('Multiply each element by 2', arr*2)
print('filter elements > 25', arr[arr>25])
data={'Name':['Alice','Bob','Charlie','David'],'Age':[25,30,35,40],'Score':[85.5,92.0,88.0,76.5]}
df=pd.DataFrame(data)
print("\n Original DataFrame:\n ",df)
df['Passed']=df['Score']>=80
print("\nDataFrame with 'Passed' column:\n",df)
older_students=df[df['Age']>30]
print("\n Students older than 30:\n",older_students)
average_score=df['Score'].mean()
print("\nAverage Score: ",average_score)
```

```python
import numpy as np
import matplotlib.pyplot as plt
temps_celsius = np.array([15.2, 17.5, 22.1, 28.0, 32.5, 34.8, 33.9, 32.0, 29.0, 24.5, 19.8, 16.0])
months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun","Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
avg_temp = np.mean(temps_celsius)
hottest_index = np.argmax(temps_celsius)
coldest_index = np.argmin(temps_celsius)
hottest_month = months[hottest_index]
coldest_month = months[coldest_index]
temps_normalized = (temps_celsius - np.min(temps_celsius)) / (np.max(temps_celsius) - np.min(temps_celsius))
above_avg_months = months[temps_celsius > avg_temp]
temps_fahrenheit = temps_celsius * 9/5 + 32
print("Average Temperature (°C):", round(avg_temp, 2))
print("Hottest Month:", hottest_month, "(", temps_celsius[hottest_index], "°C )")
print("Coldest Month:", coldest_month, "(", temps_celsius[coldest_index], "°C )")
print("Normalized Temperatures:", np.round(temps_normalized, 2))
print("Months Above Average:", above_avg_months)
print("Temperatures in Fahrenheit:", np.round(temps_fahrenheit, 2))
plt.figure(figsize=(8, 5))
plt.plot(months, temps_celsius, marker='o', label="Temperature (°C)")
plt.axhline(avg_temp, color='r', linestyle='--', label=f"Avg Temp: {avg_temp:.2f}°C")
plt.title("Monthly Temperature Trend")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
data = {"Order ID": [101, 102, 103, 104, 105, 106, 107],"Product": ["Laptop", "Smartphone", "Tablet", "Laptop", "Headphones", "Smartphone", "Tablet"],"Quantity": [2, 5, 3, 1, 10, 2, 4],"Price": [55000, 15000, 20000, 55000, 2000, 15000, 20000]}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df, "\n")
df["Total Sale"] = df["Quantity"] * df["Price"]
print("DataFrame with Total Sale:")
print(df, "\n")
high_sales_df = df[df["Total Sale"] > 1000]
print("Orders with Total Sale > 1000:")
print(high_sales_df, "\n")
product_sales = df.groupby("Product")["Quantity"].sum()
most_sold_product = product_sales.idxmax()
print("Most Sold Product:", most_sold_product, "\n")
def discount_rule(sale):
    if sale > 50000:
        return 0.10
    elif sale > 20000:
        return 0.05
    else:
        return 0.02
df["Discount Rate"] = df["Total Sale"].apply(discount_rule)
df["Discount Amount"] = df["Total Sale"] * df["Discount Rate"]
print("DataFrame with Discount Information:")
print(df, "\n")
df.to_excel("ecommerce_sales_summary.xlsx", index=False)
print("Excel file 'ecommerce_sales_summary.xlsx' created successfully.")
```
